```python
import pandas as pd
import numpy as np
import ast
import json
import re
from collections import defaultdict
```

```python
df = pd.read_csv('data/final_dataset_reduced.csv')
```

```python
dataset_dict_list = df.to_dict('records')
```

```python
dataset_dict_list[1]
```

```python
albums_to_song_indexes = defaultdict(list)

for i, song in enumerate(dataset_dict_list):
  album = song['album']
  artist = song['artist']
  if pd.isna(album) or not album:
    continue
  albums_to_song_indexes[str(album) + " - " + str(artist)].append(i)
```

```python
albums_to_song_indexes
```

```python
with open('data/albums_to_song_indexes.json', 'w') as fp:
    json.dump(albums_to_song_indexes, fp)
```
